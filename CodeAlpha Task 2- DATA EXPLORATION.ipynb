```python
#Import all the necessary libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Iris dataset
iris = pd.read_csv("Iris.csv", 
                   header=None, 
                   names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'Species'])
```

```python
#Shows all the columns
print(iris.columns)
```

```python
# Check for null values
null_values = iris.isnull().sum()
print(null_values)
```

```python
# Check the data types of the columns
print("Data types:", iris.dtypes)
```

```python
#Check the summary statistics of the numerical columns
print("Summary statistics:\n", iris.describe())
```

```python
#Create a scatter plot of the 'sepal_length' vs 'sepal_width' features, coloring the points based on the 'species' column
plt.figure(figsize=(14, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='Species')
# Set the title, labels for the plot
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
#Display the scatter plot
plt.show()
```

```python
#Create a scatter plot of the 'petal_length' vs 'petal_width' features, coloring the points based on the 'species' column
plt.figure(figsize=(13, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='Species')
# Set the title, labels for the plot
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
#Display the scatter plot
plt.show()
```

```python
# Create a histogram of the 'sepal_width' feature
plt.figure(figsize=(20, 6))
sns.histplot(data=iris, x='sepal_width', hue='Species', element='step', stat='percent')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Percent')
plt.show()
```

```python
# Create a histogram of the 'sepal_length' feature
plt.figure(figsize=(20, 6))
sns.histplot(data=iris, x='sepal_length', hue='Species', element='step', stat='percent')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Percent')
plt.show()
```
